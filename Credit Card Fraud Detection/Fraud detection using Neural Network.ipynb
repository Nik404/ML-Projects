```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./creditcard.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
df['Amount_Normal'] = sc.fit_transform(df['Amount'].values.reshape(-1,1))
```

```python
df.head()
```

```python
df = df.drop(labels = 'Amount',axis=1)
```

```python
df
```

```python
titles = list(df.columns)
```

```python
titles[-1],titles[-2] = titles[-2],titles[-1]
```

```python
df = df[titles]
```

```python
df.head()
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 0)
```

```python
import keras 

from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(16, kernel_initializer = 'uniform',activation = 'relu', input_dim = 30))
model.add(Dense(16,activation = 'relu')) #hidden layer
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
hist = model.fit(x,y,batch_size=128, epochs = 100,validation_data=(x_train,y_train))
```

```python
y_predict = np.array(model.predict(x_test))
```

```python
y_predict = (y_predict > 0.5)
print(y_predict)
```

```python
model.evaluate(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test,y_predict)
```

```python
print(f"Accuracy of the model: {accuracy_score(y_test,y_predict)*100}%")
```
